```python
import pandas as pd
import torch
import numpy as np
from scipy.special import log_softmax
from expected_cost.utils import plot_hists
```

```python
seed = 0
results_dir = f"../outputs/adaptation/llama3.2-1b/lora_ans_instruct_plus_dp_cal_alltrainsamples/sst2_128_0.3_{seed}/test=sst2/list=test_400"
logits = pd.read_csv(f"{results_dir}/logits.csv", index_col=0, header=None).values.astype(float)
logprobs = log_softmax(logits, axis=1)
labels = pd.read_csv(f"{results_dir}/labels.csv", index_col=0, header=None).values.astype(int).flatten()

plot_hists(labels, logprobs, nbins=100, group_by="score")
```
